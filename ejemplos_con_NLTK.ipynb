```python
print('Distribución de Frequencia')
fdist1 = nltk.FreqDist(text1)
print(fdist1)

print(fdist1['whole'])
print(fdist1['love'])
```

```python
print('Distribución de Frequencia')
fdist2 = nltk.FreqDist(text2)
print(fdist2)

print(fdist2['badness'])
print(fdist2['Elinor'])
```
