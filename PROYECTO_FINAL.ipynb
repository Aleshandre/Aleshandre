# # **PREGUNTA 1**

a) Crear una funcin que reciba como entrada un capital, un tipo de inters y un nmero de aos, y muestre por pantalla el valor futuro de la inversin cada ao del periodo indicado.

```python
def valor_futuro(vp, i, n):
  return vp * (1 + i) ** n

def valor_presente(vf, i, n):
  return vf / (1 + i) ** n

opcin = input("Desea calcular el valor futuro (F) o el valor presente (P)? ")

if opcin.upper() == "F":
  vp = float(input("Ingrese el valor presente: "))
  i = float(input("Ingrese la tasa de inters: "))
  n = int(input("Ingrese el nmero de perodos: "))

  for ao in range(1, n + 1):
    vf = valor_futuro(vp, i, ao)
    print(f"El valor futuro en el ao {ao} es: {vf:.2f}")

elif opcin.upper() == "P":
  vf = float(input("Ingrese el valor futuro: "))
  i = float(input("Ingrese la tasa de inters: "))
  n = int(input("Ingrese el nmero de perodos: "))

  for ao in range(1, n + 1):
    vp = valor_presente(vf, i, ao)
    print(f"El valor presente en el ao {ao} es: {vp:.2f}")

else:
  print("Opcin invlida.")
```

b) Crear una funcin que reciba como entrada un capital, un tipo de inters y un nmero de aos, y muestre por pantalla el valor actual del capital cada ao del periodo indicado.

```python
import matplotlib.pyplot as plt

def valor_futuro(vp, i, n):
  return vp * (1 + i) ** n

def valor_presente(vf, i, n):
  return vf / (1 + i) ** n

opcin = input("Desea calcular el valor futuro (F) o el valor presente (P)? ")

if opcin.upper() == "F":
  vp = float(input("Ingrese el valor presente: "))
  i = float(input("Ingrese la tasa de inters: "))
  n = int(input("Ingrese el nmero de perodos: "))

  valores_futuros = []
  for ao in range(1, n + 1):
    vf = valor_futuro(vp, i, ao)
    valores_futuros.append(vf)

  plt.plot(range(1, n + 1), valores_futuros)
  plt.xlabel("Aos")
  plt.ylabel("Valor futuro")
  plt.show()

elif opcin.upper() == "P":
  vf = float(input("Ingrese el valor futuro: "))
  i = float(input("Ingrese la tasa de inters: "))
  n = int(input("Ingrese el nmero de perodos: "))

  valores_presentes = []
  for ao in range(1, n + 1):
    vp = valor_presente(vf, i, ao)
    valores_presentes.append(vp)

  plt.plot(range(1, n + 1), valores_presentes)
  plt.xlabel("Aos")
  plt.ylabel("Valor presente")
  plt.show()

else:
  print("Opcin invlida.")
```

### 6.   **Mostrar por pantalla el precio medio en euros de los coches agrupando por marca y ordenando de menor a mayor precio.**

```python
import pandas as pd

# Cargar el DataFrame con la informacin de los modelos de carros vendidos
df = pd.read_csv('coches.csv',sep=";")

# Eliminar las filas con valores perdidos
df_filtered=df.dropna()

# Agrupamos los coches por marca y calculamos el precio medio de cada marca
precio_medio = df.groupby('Marca')['Precio'].mean()

# Ordenamos los resultados de menor a mayor precio
precio_medio = precio_medio.sort_values()

# Mostramos por pantalla el precio medio de cada marca
print(precio_medio)
```

### 7.   **Dibujar el diagrama de barras del porcentaje de modelos de cada marca.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame con la informacin de los modelos de carros vendidos
df = pd.read_csv('coches.csv',sep=";")

# Eliminar las filas con valores perdidos
df_filtered=df.dropna()

# Agrupamos los coches por marca y contamos el nmero de modelos de cada marca
modelos_por_marca = df.groupby('Marca')['Modelo'].count()

# Calculamos el porcentaje de modelos de cada marca sobre el total de modelos
porcentaje_modelos = modelos_por_marca / modelos_por_marca.sum()

# Dibujamos el diagrama de barras
porcentaje_modelos.plot.bar()

# Establecemos el ttulo del grfico
plt.title('Modelos mas vendidos por Marca')

# Establecemos el ttulo del eje X
plt.xlabel('Marcas')

# Establecemos el ttulo del eje Y
plt.ylabel('Porcentaje')

# Mostramos el grfico
plt.show()
```
