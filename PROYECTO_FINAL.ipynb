# # **PREGUNTA 1**

a) Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor futuro de la inversión cada año del periodo indicado.

```python
def valor_futuro(vp, i, n):
  return vp * (1 + i) ** n

def valor_presente(vf, i, n):
  return vf / (1 + i) ** n

opción = input("¿Desea calcular el valor futuro (F) o el valor presente (P)? ")

if opción.upper() == "F":
  vp = float(input("Ingrese el valor presente: "))
  i = float(input("Ingrese la tasa de interés: "))
  n = int(input("Ingrese el número de períodos: "))

  for año in range(1, n + 1):
    vf = valor_futuro(vp, i, año)
    print(f"El valor futuro en el año {año} es: {vf:.2f}")

elif opción.upper() == "P":
  vf = float(input("Ingrese el valor futuro: "))
  i = float(input("Ingrese la tasa de interés: "))
  n = int(input("Ingrese el número de períodos: "))

  for año in range(1, n + 1):
    vp = valor_presente(vf, i, año)
    print(f"El valor presente en el año {año} es: {vp:.2f}")

else:
  print("Opción inválida.")
```

b) Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor actual del capital cada año del periodo indicado.

```python
import matplotlib.pyplot as plt

def valor_futuro(vp, i, n):
  return vp * (1 + i) ** n

def valor_presente(vf, i, n):
  return vf / (1 + i) ** n

opción = input("¿Desea calcular el valor futuro (F) o el valor presente (P)? ")

if opción.upper() == "F":
  vp = float(input("Ingrese el valor presente: "))
  i = float(input("Ingrese la tasa de interés: "))
  n = int(input("Ingrese el número de períodos: "))

  valores_futuros = []
  for año in range(1, n + 1):
    vf = valor_futuro(vp, i, año)
    valores_futuros.append(vf)

  plt.plot(range(1, n + 1), valores_futuros)
  plt.xlabel("Años")
  plt.ylabel("Valor futuro")
  plt.show()

elif opción.upper() == "P":
  vf = float(input("Ingrese el valor futuro: "))
  i = float(input("Ingrese la tasa de interés: "))
  n = int(input("Ingrese el número de períodos: "))

  valores_presentes = []
  for año in range(1, n + 1):
    vp = valor_presente(vf, i, año)
    valores_presentes.append(vp)

  plt.plot(range(1, n + 1), valores_presentes)
  plt.xlabel("Años")
  plt.ylabel("Valor presente")
  plt.show()

else:
  print("Opción inválida.")
```

### 5.   **Mostrar por pantalla la marca y el modelo del coche más caro.**

```python
import pandas as pd

# Cargar el DataFrame con la información de los modelos de carros vendidos
df = pd.read_csv('coches.csv',sep=";")

# Eliminar las filas con valores perdidos
df_filtered=df.dropna()

# Seleccionamos la columna de precio y obtenemos el precio más alto
precio_maximo = df['Precio'].max()

# Obtenemos la fila que tiene el precio más alto
fila_maxima = df['Precio'].idxmax()

# Accedemos a la marca y el modelo del coche más caro usando la fila seleccionada
marca_modelo = df.loc[fila_maxima, ['Marca', 'Modelo']]

# Mostramos por pantalla la marca y el modelo del coche más caro
print(marca_modelo)
```

### 6.   **Mostrar por pantalla el precio medio en euros de los coches agrupando por marca y ordenando de menor a mayor precio.**

```python
import pandas as pd

# Cargar el DataFrame con la información de los modelos de carros vendidos
df = pd.read_csv('coches.csv',sep=";")

# Eliminar las filas con valores perdidos
df_filtered=df.dropna()

# Agrupamos los coches por marca y calculamos el precio medio de cada marca
precio_medio = df.groupby('Marca')['Precio'].mean()

# Ordenamos los resultados de menor a mayor precio
precio_medio = precio_medio.sort_values()

# Mostramos por pantalla el precio medio de cada marca
print(precio_medio)
```

### 7.   **Dibujar el diagrama de barras del porcentaje de modelos de cada marca.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame con la información de los modelos de carros vendidos
df = pd.read_csv('coches.csv',sep=";")

# Eliminar las filas con valores perdidos
df_filtered=df.dropna()

# Agrupamos los coches por marca y contamos el número de modelos de cada marca
modelos_por_marca = df.groupby('Marca')['Modelo'].count()

# Calculamos el porcentaje de modelos de cada marca sobre el total de modelos
porcentaje_modelos = modelos_por_marca / modelos_por_marca.sum()

# Dibujamos el diagrama de barras
porcentaje_modelos.plot.bar()

# Establecemos el título del gráfico
plt.title('Modelos mas vendidos por Marca')

# Establecemos el título del eje X
plt.xlabel('Marcas')

# Establecemos el título del eje Y
plt.ylabel('Porcentaje')

# Mostramos el gráfico
plt.show()
```
